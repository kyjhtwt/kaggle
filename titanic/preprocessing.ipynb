```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout
import os
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import tensorflow.keras.layers as layers
import yaml
```

```python
with open('config.yml') as file:
    config = yaml.load(file, Loader=yaml.FullLoader)
test = config['system']['test']
```

```python
if(test):
    preprocessing_file = 'test.csv'
else:
    preprocessing_file = 'train.csv'
```

```python
def get_age_range(x):
    if(x <= 11): return 0
    else: return 1
```

```python
dataset = pd.read_csv(preprocessing_file)
```

```python
dataset = dataset.drop(columns=['Cabin', 'Ticket', 'Name'])
```

```python
temp_df = dataset.copy()
temp_df = temp_df.fillna(-1)
condition = temp_df['Age'] >= 0
temp_df = temp_df[condition]
age_sum = temp_df['Age'].sum(axis=0)
valid_age_cnt = temp_df['Age'].count()
age_avg = round(age_sum / valid_age_cnt)
dataset['Age'] = dataset['Age'].fillna(age_avg)
```

```python
dataset['is_infant'] = dataset['Age'].apply(lambda x : get_age_range(x))
dataset['family'] = (dataset['Parch'] + dataset['SibSp'])
```

```python
if(test):
    dataset = dataset.fillna(30)
```

```python
if(test):
    categorical_columns_onehotEncoding = ['Sex', 'Embarked']
    categorical_columns_labelEncoding = ['Pclass']
    for col in categorical_columns_labelEncoding:
        le = preprocessing.LabelEncoder()
        le.fit(dataset[col].values)
        dataset[col] = le.transform(dataset[col].values)
    dataset = pd.get_dummies(data=dataset, columns = categorical_columns_onehotEncoding)
    dataset = dataset.drop(columns=col)
    dataset = dataset.drop(columns=['Sex_male', 'Embarked_S', 'Embarked_Q'])
    dataset.to_excel('test.xlsx', index=False)
else:
    categorical_columns_onehotEncoding = ['Sex', 'Embarked']
    categorical_columns_labelEncoding = ['Pclass']
    for col in categorical_columns_labelEncoding:
        le = preprocessing.LabelEncoder()
        le.fit(dataset[col].values)
        dataset[col] = le.transform(dataset[col].values)
    dataset = dataset.drop(columns='PassengerId')
    dataset = pd.get_dummies(data=dataset, columns = categorical_columns_onehotEncoding)
    dataset = dataset.drop(columns=col)
    dataset = dataset.drop(columns=['Sex_male', 'Embarked_S', 'Embarked_Q'])
    dataset.to_excel('preprocessing.xlsx', index=False)
```
